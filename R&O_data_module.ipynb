```python
import os
import requests

repo_owner = "wadoran"
repo_name = "Excel-files-Risks"
download_dir = "excel_files"

# Create the directory if it doesn't exist
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Get the list of repository contents using the GitHub API
api_url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/contents"
response = requests.get(api_url)
data = response.json()

# Iterate over the contents and download Excel files
for item in data:
    if item["type"] == "file" and item["name"].endswith(".xlsx"):
        file_url = item["download_url"]
        file_name = item["name"]
        file_path = os.path.join(download_dir, file_name)
        
        # Send a request to download the file
        response = requests.get(file_url)
        
        # Save the file to disk
        with open(file_path, "wb") as file:
            file.write(response.content)
            
        print(f"Downloaded: {file_name}")
```
